# 1. User Login Check

```python
username , password = "admin" , "1234"
if username == "admin" and password == "1234":
    print("Login Successful")
else:
    print("Invalid Credentials")
```

# 2. Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
passed, failed = 0,0
for i in range(marks.__len__()):
  if marks[i] >= 50:
    passed += 1
  else:
    failed += 1
print("Passed students: ", passed)
print("Failed students: ", failed)
```

# 3. Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
Names = []
for name in names:
    Names.append(name.strip().lower())
print(Names)
```

# 4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
flagged_messages = []
for msg in messages:
  print(f"Message: {msg} Length: {len(msg)}")
  if len(msg) > 10:
    flagged_messages.append(msg)
    print(f"Flagged Messages: {flagged_messages}")
```

# 5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count = 0
for log in logs:
  if log == "ERROR":
    count += 1
print(f"Number of errors: {count}")
```
