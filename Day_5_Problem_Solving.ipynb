# **Problem 1: Social Media – Post Engagement Analyzer**

```python
def post_engagement_analyzer(likes_list):
  total_likes = 0
  for likes in likes_list:
    total_likes += likes
  print("Total Likes : ", total_likes)
  if total_likes >= 1000:
    print("Post Status : Viral Post")
  else:
    print("Post Status : Normal Engagement")
likes_list = [200, 150, 300, 400]
post_engagement_analyzer(likes_list)
#
```

# **Problem 3: Agriculture – Rainfall Adequacy Checker**

```python
def rainfall_adequacy_checker(rainfall_list,required_rainfall):
  total_rainfall = 0
  for rainfall in rainfall_list:
    total_rainfall += rainfall
  average_rainfall = total_rainfall / len(rainfall_list)
  print("Average Rainfall :", int(average_rainfall))
  if average_rainfall >= required_rainfall:
    print("Status: Adequate rainfall")
  else:
    print("Status: Inadequate rainfall")
rainfall_data = [70, 75, 80, 65, 70]
required = 70
rainfall_adequacy_checker(rainfall_data, required)
```

# **Problem 4: Social Media – Duplicate Account Detection**

```python
def duplicate_account_detection(usernames):
  username_count = {}
  for username in usernames:
    if username in username_count:
      username_count[username] += 1
    else:
      username_count[username] = 1
  for username, count in username_count.items():
    if count > 1:
      print("Duplicate Accounts Found : Yes")
      return
  print("No Duplicate Accounts Found : No")
user_list = ["nikil", "rahul", "ananya", "rahul"]
duplicate_account_detection(user_list)
```

# **Problem 6: Agriculture – Premium Crop Price Filter**

```python
def filter_premium_crops(prices):
    premium_crops = []
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    print("Premium Crops:", premium_crops)
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

# **Problem 9: E-Learning – Student Attendance Eligibility System**

```python
def check_attendance(attendance_list):
    total_classes = len(attendance_list)
    attended_classes = 0
    for status in attendance_list:
        if status == 1:
            attended_classes += 1
    percentage = (attended_classes / total_classes) * 100
    print("Attendance Percentage:", percentage)
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
attendance_data = [1, 1, 0, 1, 1]
check_attendance(attendance_data)
```

# **Problem 10: Smart Electricity Bill Analyzer**

```python
def calculate_electricity_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    return bill, usage_status
units_consumed = 250
total_bill, status = calculate_electricity_bill(units_consumed)
print("Units Consumed:", units_consumed)
print("Total Bill:", total_bill)
print("Usage Status:", status)
```
