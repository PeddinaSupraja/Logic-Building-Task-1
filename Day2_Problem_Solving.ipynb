# **1.Unique Words in a Sentence**

```python
sentence=input ("Enter the Sentence:")
words=sentence.split()
unique_words=set(words)
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

# **2.Highest Salary from Employee Data**

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
name = max(employees, key=employees.get)
salary = employees[name]
print(f"Highest Salary: {name} - {salary}")
```

# **3.Find Maximum and Minimum Values**

```python
numbers = [45, 22, 89, 10, 66]
max_val = numbers[0]
min_val = numbers[0]
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num
print("List:", numbers)
print("Maximum:", max_val)
print("Minimum:", min_val)
```

# **4.Count Products Above a Price Threshold**

```python
prices = [450, 1200, 899, 1500, 3000]
count = 0
for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000:", count)
```

# **5.Calculate Attendance Percentage**

```python
attendance = ["P", "P", "A", "P", "P"]
present_count = attendance.count("P")
total_days = len(attendance)
attendance_percentage = (present_count / total_days) * 100
print("Attendance Percentage:", attendance_percentage)
```

# **6.Remove Duplicate Phone Numbers**

```python
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_numbers = set(phone_numbers)
print("Unique phone numbers:", unique_numbers)
```

## **7.Count Character Frequency**

```python
text = "python"
char_count = {}
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1
print(char_count)
```

# **9: Check if a Key Exists in a Dictionary**

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
key_to_check = "Ravi"
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee not found")
```
