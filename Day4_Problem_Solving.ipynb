# **Problem 1: Mobile Recharge Validation System**

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount < 50:
        print(" Recharge amount must be at least ₹50.")
        return False
    elif amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans: 199, 299, 399, 599")
        return False
    else:
        print("Recharge successful!")
        return True
# Main Program
while True:
    try:
        recharge_amount = int(input("Enter recharge amount: ₹"))
        # Call function
        if validate_recharge(recharge_amount):
            break   # Exit loop if valid
        print("Please try again.\n")
    except ValueError:
        print("Please enter a valid number.\n")
```

# **Problem 2: Inventory Reorder Alert System**

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"Reorder Alert: {product} stock is low ({stock} items left)")
        else:
            print(f"Stock OK: {product} has sufficient stock ({stock} items)")
# Dictionary of products and their stock quantities
inventory = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 12,
    "Soap": 30,
    "Milk Powder": 8
}
# Function call
check_inventory(inventory)
```

# **Problem 3: Student Result Processing System**

```python
def process_result(marks):
    total = 0
    # Loop to calculate total marks
    for mark in marks:
        total += mark
    average = total / len(marks)
    #  To Determine the result
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"
    return average, result
# Example list of marks
student_marks = [60, 45, 70, 55, 40]
# Function call
avg, status = process_result(student_marks)
print("Average Marks:", avg)
print("Result:", status)
```

# **Problem 4: Cab Fare Estimator with Retry Option**

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    total_fare = base_fare + (distance * per_km_rate)
    # Apply peak hour charges
    if peak.lower() == "yes":
        total_fare += total_fare * 0.25   # 25% extra
    return total_fare
# Main Program with retry option
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_hour = input("Is it peak hour? (yes/no): ")
        fare = calculate_fare(distance, peak_hour)
        print(f"Total Cab Fare: ₹{fare:.2f}")
        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            print("Thank you for using Cab Fare Estimator!")
            break
    except ValueError:
        print("Please enter valid numeric distance.\n")
```

# **Problem 5: Employee Attendance Eligibility Checker**

```python
def check_attendance(attendance_list):
    present_count = 0
    # Loop to count present days
    for day in attendance_list:
        if day.upper() == "P":
            present_count += 1
    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100
    # Determine eligibility
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    return percentage, status
# Example attendance list
attendance = ["P", "P", "A", "P", "A", "P", "P", "P"]
# Function call
percent, result = check_attendance(attendance)
print(f"Attendance Percentage: {percent:.2f}%")
print("Eligibility Status:", result)
```

# **Problem 6: Password Strength Checker**

```python
def check_password(password):
    has_digit = False
    has_special = False
    # Check minimum length
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    # Loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True
    # Check conditions
    if has_digit and has_special:
        return "Strong Password "
    else:
        return "Weak Password (Must contain at least one digit and one special character @#$)"
# Example usage
user_password = input("Enter your password: ")
result = check_password(user_password)
print("Password Strength:", result)
```
