# **Problem 1: Smart Light Controller**

```python
class SmartLightController:
  def __init__(self,name):
    self.name = name
    self.status = "OFF"
  def turn_on(self):
    self.status = "ON"
  def turn_off(self):
    self.status = "OFF"
  def show_status(self):
    print(f"{self.name} is {self.status}")
light = SmartLightController("Bedroom Light")
light.turn_on()
light.show_status()
```

# **Problem 2: Employee ID Card System**

```python
class EmployeeIDCard:
  def __init__(self,name,emp_id,department):
    self.name = name
    self.emp_id = emp_id
    self.department = department
  def display_info(self):
    print("Employee ID Card")
    print("Name:",self.name)
    print("ID:",self.emp_id)
    print("Department:",self.department)
emp = EmployeeIDCard("Rahul", "EMP102", "IT")
emp.display_info()
```

# **Problem 3: Mobile Contact Record**

```python
class MobileContactRecord:
  def __init__(self,name,contact):
    self.name = name
    self.contact = contact
  def display_info(self):
    print("Contact Saved")
    print("Name:",self.name)
    print("Contact:",self.contact)
c1 = MobileContactRecord("Anita", 9876543210)
c1.display_info()
```

# **Problem 4: Product Price Tag Generator**

```python
class ProductPriceTag:
  def __init__(self,product,price):
    self.product = product
    self.price = price
  def display_info(self):
    print("Product:",self.product)
    print("Price: ₹"+str(self.price))
p = ProductPriceTag("Headphones",2499 )
p.display_info()
```

# **Problem 5: Movie Rating Display System**

```python
class MovieRatingDisplay:
  def __init__(self,movie,rating):
    self.movie = movie
    self.rating = rating
  def display_info(self):
    print("Movie:",self.movie)
    print("Rating:",self.rating,"/ 5")
m = MovieRatingDisplay("Inception", 4.8)
m.display_info()
```

# **Problem 6: Delivery Address Manager**

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    def display_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)
d1 = Delivery("Suman", "Hyderabad")
d1.display_details()
```
